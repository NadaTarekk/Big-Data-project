It is in the object type, we need to change it into the int type

```python
df['Installs'] = df['Installs'].str.split('+').str[0]    # reomoves the + symbol
df['Installs'].replace(',','', regex=True, inplace=True) #replace the commas ',' in the install column
df['Installs'] = df['Installs'].astype(int)              # converting it to the int type
```

As the size of data can be in GB, MB and KB so we will convert the data into the size in MB

```python
df['Size'] = df['Size'].apply(lambda x: str(x).replace('M', '') if 'M' in str(x) else x)
```

```python
df['Size'] = df['Size'].apply(lambda x: str(x).replace(',', '.') if ',' in str(x) else x)
```

conversion of kbs to mb

```python
df['Size'] = df['Size'].apply(lambda x: float(str(x).replace('k', '')) / 1000 if 'k' in str(x) else x)
```

there is missmatch value in the data set which is 'Varies with device' it means the size of the App may vaies with the device---> but we will assume it as 0

```python
df['Size'] = df['Size'].apply(lambda x: str(x).replace('Varies with device', '0') if 'Varies with device' in str(x) else x)
```

conversion of data Gbs to Mb

```python
df['Size'] = df['Size'].apply(lambda x: float(str(x).replace('G', '')) * 1000 if 'G' in str(x) else x)
df['Size'] = df['Size'].apply(lambda x: float(x))
df.dtypes['Size']
```

##1) What are the top 10 Categories that are installed from the Google Play Store ?

```python
df['Category'].unique()
```

DataFrame top_Category counts the occurrences of each category in the Category column then resets the index of the resulting Series to make it a DataFrame, renames the columns, and stores the result in top_Category.

```python
top_Category = df['Category'].value_counts().reset_index()
top_Category.columns = ['Category', 'Count']
```

View top_category

```python
top_Category
```

Group category_installs df by the Category column and calculates the sum of the Installs column for each category

```python
Category_installs = df.groupby(['Category'])[['Installs']].sum()
```

```python
Category_installs
```

Merge DataFrames top_Category and Category_installs based on the common column 'Category' to know number of installs for each category

```python
top_Category_installs = pd.merge(top_Category, Category_installs, on='Category')
top_Category_installs.head(5)
```

```python
top_10_Categories_installs = top_Category_installs.head(10).sort_values(by = ['Installs'],ascending = False)
```

```python
plt.figure(figsize=(14,7))
plt.xticks(rotation=60)
plt.xlabel("Category")
plt.ylabel("Number of applications")
plt.title("Top 10 Installed Categories")
sns.barplot(x = top_10_Categories_installs.Category,y= top_10_Categories_installs.Installs)
```

## 4) Which Category has the highest Paid and Free apps ?

group and aggregate your DataFrame to get the counts of apps by category and type then use the pivot() function to reshape the DataFrame based on the specified columns.

```python
app_count = df.groupby(['Category', 'Type'])[['App Name']].count().reset_index().rename(columns={'App Name': 'Count'})
df_app_count = app_count.pivot(index='Category', columns='Type', values='Count').fillna(0).reset_index()
```

```python

df_app_count.set_index('Category').plot(kind='bar', stacked=True, figsize=(18,9)) # try df_app_count.plot(kind='bar', stacked=True, figsize=(18,9))
plt.xlabel("Category", fontsize=15)
plt.ylabel("Count", fontsize=15)
plt.title("Free and Paid Apps in Category")
```

<font size="5">Answer is Education Category is the highest in paid apps list</font>

<font size="5">Observation :</font>
It looks like certain app categories have more free apps available for download than others. In our dataset, the majority of apps in Business,Education, Music & Audio, Tools as well as Social categories were free to install. There is a high variance in the number of installs, we need to reduce it so we can use a log value for this column, otherwise it would be unable to see the data when we visualize

```python
categoty_type_installs = df.groupby(['Category','Type'])[['Installs']].sum().reset_index()
categoty_type_installs['log_Installs'] = np.log10(categoty_type_installs['Installs'])
```

```python
plt.figure(figsize=(18,9))
plt.xticks(rotation=65,fontsize=9)
plt.xlabel("Category")
plt.ylabel("Installs")
plt.title("Number of installed Apps type wise according to Category")
sns.barplot(x = 'Category', y = 'log_Installs', hue='Type', data=categoty_type_installs)
```
